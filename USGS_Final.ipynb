```python
filenames= ["usco2015Edit.xlsx","usco2010Edit.xlsx","usco2005Edit.xls","usco2000Edit.xls","usco1995Edit.xlsx","usco1990Edit.xls","usco1985Edit.xls"]


Water_data_list=[]
for filename in filenames: 
    temp_df= pd.read_excel(filename, header=0)
    temp_df["filename"]=filename 
    temp_df.columns = [x.lower() for x in temp_df.columns]
    display(temp_df.head())
    Water_data_list.append(temp_df)

Water_data_years=pd.concat(Water_data_list, axis=0, ignore_index=True)

Water_data_years.to_excel("Water_data_years4.xlsx", index=False)
```
