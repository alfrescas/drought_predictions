```python
#2015 sort by State

test_2015 = pd.DataFrame()

State_val = ['WY', 'UT', 'NM', 'CO']

for val in State_val:
    test_2015=test_2015.append(DF_2015.loc[DF_2015['STATE'] == (val)])
    
#2015 rename statefips/countyfips/total population/groundwater population
test_2015.rename(columns = {'STATEFIPS': 'scode', 'COUNTYFIPS':'CountyCode', 'TP-TotPop':'TotalPop'}, inplace = True)
    
test_2015.head()
```

```python
#2010 sort by state
test_2010 = pd.DataFrame()

State_val = ['WY', 'UT', 'NM', 'CO']

for val in State_val:
    test_2010=test_2010.append(DF_2010.loc[DF_2010['STATE'] == (val)])
    
#2010 rename statefips/countyfips/total population
test_2010.rename(columns = {'STATEFIPS': 'scode', 'COUNTYFIPS':'CountyCode', 'TP-TotPop':'TotalPop'}, inplace = True)
    
test_2010.head()
```

```python
#2005 sort by state
test_2005 = pd.DataFrame()

State_val = ['WY', 'UT', 'NM', 'CO']

for val in State_val:
    test_2005=test_2005.append(DF_2005.loc[DF_2005['STATE'] == (val)])
    
#2005 rename statefips/countyfips/total population
test_2005.rename(columns = {'STATEFIPS': 'scode', 'COUNTYFIPS':'CountyCode', 'TP-TotPop':'TotalPop'}, inplace = True)

#add YEAR column to df
test_2005 = test_2005.assign(YEAR = '2005')
    
test_2005.head()
```

```python
#2000 sort by state
test_2000 = pd.DataFrame()

State_val = ['WY', 'UT', 'NM', 'CO']

for val in State_val:
    test_2000=test_2000.append(DF_2000.loc[DF_2000['STATE'] == (val)])
    
#2000 rename statefips/countyfips/total population
test_2000.rename(columns = {'STATEFIPS': 'scode', 'COUNTYFIPS':'CountyCode', 'TP-TotPop':'TotalPop'}, inplace = True)

### DOESN"T CONTAIN A PS-GWPOP COLUMN

#add YEAR to column
test_2000 = test_2000.assign(YEAR = '2000')

test_2000.head()
```

```python
#1995 sort by state
test_1995 = pd.DataFrame()

State_val = ['WY', 'UT', 'NM', 'CO']

for val in State_val:
    test_1995=test_1995.append(DF_1995.loc[DF_1995['State'] == (val)])
    
#1995 rename state/year/state code column
test_1995.rename(columns = {'State':'STATE', 'Year':'YEAR', 'StateCode':'scode'}, inplace = True)
    
test_1995.head()
```

```python
#1990 sort by state
test_1990 = pd.DataFrame()

State_val = ['WY', 'UT', 'NM', 'CO']

for val in State_val:
    test_1990=test_1990.append(DF_1990.loc[DF_1990['state'] == (val)])
    
#1990 rename state/year/area column/total population/public supply total population served/pub supply gw total/pub supply GW withdrawals fresh
test_1990.rename(columns = {'state':'STATE', 'year':'YEAR', 'area':'CountyCode', 'po-total':'TotalPop', 'ps-popto':'PS-TOPop', 'ps-popgw':'PS-GWPop', 'ps-wgwfr': 'PS-WGWFr'}, inplace = True)
    
test_1990.head()
```

```python
#1985 sort by state
test_1985 = pd.DataFrame()

State_val = ['WY', 'UT', 'NM', 'CO']

for val in State_val:
    test_1985=test_1985.append(DF_1985.loc[DF_1985['state'] == (val)])
    
#1985 rename state/year/area column/population total/public supply pop total/pub supply gw total
test_1985.rename(columns = {'state':'STATE', 'year': 'YEAR', 'area':'CountyCode', 'po-total':'TotalPop', 'ps-popto':'PS-TOPop','ps-popgw':'PS-GWPop', 'ps-wgwfr': 'PS-WGWFr'}, inplace = True)
    
test_1985.head()
```

```python
test_1985.dtypes
```

```python
years_df = pd.concat([test_2015, test_2010, test_2005, test_2000, test_1995, test_1990, test_1985])
years_df
```

```python
#Export to csv file with index
years_df.to_csv("USGS_WaterData2.csv", index=True)
```

```python
#Export csv file without index
years_df.to_csv("USGS_WaterData.csv", index=False)
```
